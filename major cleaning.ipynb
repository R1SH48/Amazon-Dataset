```python
def user_count(string):
    string = str(string)
    split_list = set(string.split(','))
    result = len(split_list)
    return result
```

```python
def price_value(money):
    money = str(money)
    money = money.replace(',','')
    money = money.replace('','')
    for digit in money:
        if digit.isdigit() == False :
            money = money.replace(digit,'')
    return float(money)
```

```python
working_copy['discounted_price'] = working_copy['discounted_price'].apply(lambda y: price_value(y))
```

```python
working_copy['actual_price'] = working_copy['actual_price'].apply(lambda y: price_value(y))
```

```python
def discount_value(discount):
    discount = str(discount)
    discount = float(discount.replace('%', ''))
    return discount
```

```python
working_copy['discount_percentage'] = working_copy['discount_percentage'].apply(lambda x: discount_value(x))
```

```python
def rating_value(rating):
    #rating = str(rating)
    rating = str(rating)
    rating = rating.replace(',','')
    return float(rating)
```

```python
working_copy['rating_count'] = working_copy['rating_count'].apply(lambda x: rating_value(x))
```
