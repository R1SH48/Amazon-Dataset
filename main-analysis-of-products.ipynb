# rating record datatype change

```python
string = '&^*%(&^%)'
rating_datatype(string)
```

```python
def rating_datatype(string):
    flag = 1
    for char in string:
        if (char.isdigit() or char=='.'):
            continue
        else:
            flag = 0
    if flag:
        return float(string)
    else:
        return 0.0
```

```python
working_copy['rating'] = working_copy['rating'].apply(lambda x: rating_datatype(x))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# --------------------------
# 1. Basic Overview
# --------------------------
print("Shape of dataset:", working_copy.shape)
print("\nData Types:\n", working_copy.dtypes)
print("\nMissing Values:\n", working_copy.isnull().sum())
print("\nDuplicate Rows:", working_copy.duplicated().sum())

# --------------------------
# 2. Summary Statistics
# --------------------------
print(working_copy[['rating', 'rating_count', 'user_count']].describe())

# --------------------------
# 3. Distribution Plots
# --------------------------
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.histplot(working_copy['rating'], kde=True)
plt.title("Rating Distribution")

plt.subplot(1,3,2)
sns.histplot(working_copy['rating_count'], kde=True)
plt.title("Rating Count Distribution")

plt.subplot(1,3,3)
sns.histplot(working_copy['user_count'], kde=True)
plt.title("User Count Distribution")

plt.tight_layout()
plt.show()

# --------------------------
# 4. Correlation Matrix
# --------------------------
plt.figure(figsize=(8,6))
sns.heatmap(working_copy[['rating','rating_count','user_count']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# --------------------------
# 5. Category Analysis
# --------------------------

# Select category columns (exclude numeric)
category_cols = working_copy.columns[2:-1]  # Adjust if needed

# Count products per category
category_counts = working_copy[category_cols].sum().sort_values(ascending=False)

print("Top Categories by Product Count:")
print(category_counts.head(10))

# Plot top 10 categories
plt.figure(figsize=(10,6))
category_counts.head(10).plot(kind='bar')
plt.title("Top 10 Categories by Product Count")
plt.show()

# --------------------------
# 6. Average Rating per Category
# --------------------------
avg_rating_per_cat = {}

for col in category_cols:
    avg_rating_per_cat[col] = working_copy[working_copy[col] == 1]['rating'].mean()

avg_rating_df = pd.Series(avg_rating_per_cat).sort_values(ascending=False)

print("Top 10 Categories by Average Rating:")
print(avg_rating_df.head(10))
```
