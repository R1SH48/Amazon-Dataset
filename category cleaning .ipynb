```python
def category_split(string):
    string = str(string)
    split_values = string.split('|')
    strip_values = list(x.strip() for x in split_values)
    return strip_values
```

```python
working_copy['category'] = working_copy['category'].apply(lambda y: category_split(y))
```

```python
unique_category = []
def custom_category(input_list):
    for sub_list in input_list:
        for cat in sub_list:
            if cat not in unique_category:
                unique_category.append(cat)
    return unique_category
```
